```python
#importing some useful packages
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2

%matplotlib inline
```

```python
points = [[40, 30], [30, 35], [20, 40], [10, 45]]


cart_binary = np.zeros([50, 50])
for  [x, y] in points:
    cart_binary[x, y] = 255
    
fig=plt.figure(figsize=(18, 18))
fig.add_subplot(2, 2, 1)
plt.imshow(cart_binary, cmap='Greys_r')

rhos=[]
θ = np.arange(0, 180, 2)

fig.add_subplot(2, 2, 2)
for [x, y] in points:
    rh = []
    for theta in θ:
        rho = x*np.cos(np.deg2rad(theta)) + y*np.sin(np.deg2rad(theta))
        rh = rh + [rho]
    rhos = rhos + [rh]
    plt.plot(θ, rh, 'r-')
    
    
```
